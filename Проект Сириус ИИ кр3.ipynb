```python
import torch
from transformers import pipeline

# Инициализация пайплайна для суммаризации
# Используем предобученную модель для русского языка от Сбербанка (например, "sberbank-ai/ruT5-small")
summarizer = pipeline("summarization", model="RussianNLP/FRED-T5-Summarizer")

# Суммаризация текста
summary = summarizer(text, max_length=text_size, min_length=25, do_sample=False)

# Вывод результата
print("Суммаризация:")
print(summary[0]['summary_text'])
```
